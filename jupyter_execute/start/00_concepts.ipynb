# 基本概念

```{admonition} 需求
学习如何使用 C++ 表达一些简单概念。
```

信息的最小单位是 {term}`位` （也可称作 比特），一个位仅仅存储 `0` 或者 `1` 两个数字。大多数计算机中的基本寻址单元为 {term}`字节`（1 byte = 8 bit），与一个字符占据的空间大小相当。C++ 对内存做了抽象，将一块内存抽象为一个 {term}`对象`。对象的 {term}`值` 便是其对应的内存空间存储的一组比特值，值所代表的含义和所占据的比特数由其 {term}`类型` 决定。对象需要一个 **名字** 以方便使用，称给定类型的命名对象为 {term}`变量`。{term}`声明` 负责为程序引入一个新的名字，并指定该命名实体的类型。{term}`函数` 是一个命名的代码单元（或计算的逻辑单元），可以从程序中其他部分调用它。

C++ 提供了若干基本类型，例如：

- `bool`：布尔类型。可取 `true` 或 `false`。
- `char`：字符类型。如 `'a'`，`'\n'` 等
- `int`：整数类型。如 1，42 等
- `double`：双精度浮点数类型。如 2.3，5.0

这些基本类型都与硬件特性直接相关，其尺寸固定不变，决定了其中所能存储的值的范围。

## 标识符

**标识符** 是以字母和下划线开头，后接 0 个或者多个（大写或者小写）字母、数字或者下划线的序列。标识符也可以称作 **名字** 或者是 **变量名称**。

想要使用一个变量，需要使用 {term}`声明`：

````c++
int 苹果数量;
````

其中 `苹果数量` 便是一个变量；`int` 便是一个类型。

````c++
int hi_world;
int HiWorld;
````

````c++
int Hi World; // 错误：因为标识符不能有空格
````

```{margin}
C++ 使用符号 `//` 注解代码片段，以帮助人们理解。编译器不会编译注解内容。
```

````c++
int Hi@World; // 错误：因为标识符不能有 @
````

### 双下划线

以下划线开头或者包含连续两个下划线的标识符是为 C++ 实现预留的，你的程序中不要使用这种标识符。例如：

````c++
int _foo; // 不要这样用
int foo_boor; // 正确
int foo__bar; // 不要这样用
int foo_; // 正确
````

### 关键字

**关键字** 是 C++ 用来表示语言结构的特殊标识符。比如，类型 `int` 便是关键字。

### `sizeof`

可以使用 {term}`函数` `sizeof` 获取类型的空间大小。

````c++
sizeof(bool) // 1 个字节
````

````c++
sizeof(char) // 1 个字节
````

````c++
sizeof(int) //  4 个字节
````

````c++
sizeof(double)  // 8 个字节
````

## 初始化

通用初始化方法是使用花括号内的一组初始化器列表：

````c++
double d1 {2.3};
d1
````

当然也可以使用传统的方法，即赋值语句：

````c++
double d1 = 2.3;
d1
````

## 赋值语句

看几个赋值语句的例子：

````c++
int a = 7; // a 初始化为 7
a
````

````c++
a = 4; // a 通过 赋值语句，获得新值 4
a
````

````c++
int b = a; // b 的值为 a 的值的拷贝
b
````

````c++
b = a + 5; // b 获得新值 a+5
b
````

````c++
a = a + 7; // a 获得新值 a+7
a
````

### 复合赋值运算符

一个变量的递加（增加 1），可以简写为 `++vars`。比如：

````c++
int a = 7;
a // 查看 a 的值
````

````c++
++a; // 等价于 `a = a + 1;`
a
````

一些二元赋值运算符，可以直接修改变量的当前值：

````c++
int a = 7;
a += 7; // 表示 `a = a + 7`
a
````

````c++
a -= 7; // 表示 `a = a - 7`
a
````

````c++
a *= 2; // 表示 `a = a * 2`
a
````

````c++
a /= 2; // 表示 `a = a / 2`
a
````

````c++
a %= 2; // 表示 `a = a % 2`
a
````

在很多领域 `*=` 和 `/=` 被视为“缩放”。

## 类型转换

赋值运算、算术运算以及初始化会在基本类型间进行有意义的类型转换。

````c++
double d = 2.2; // 初始化浮点数
d
````

````c++
int t = 7; // 初始化整数
t
````

````c++
d = d + t; // 将求和结果赋值给 d
d // 可以看出此时 d 是浮点数
````

````c++
t = d * t; // 把乘积结果赋值给 t
t
````

`d * t` 是一个 `double` 值，而 `t` 是 `int`，所以最终的结果被截断为整数。
